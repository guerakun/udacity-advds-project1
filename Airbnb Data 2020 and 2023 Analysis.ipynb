```python
import pandas as pd
import numpy as np
# Read the first CSV
df2020 = pd.read_csv("AB_US_2020.csv", low_memory=False)
print("df2020 shape:", df2020.shape)
display(df2020.head())

# Read the second CSV
df2023 = pd.read_csv("AB_US_2023.csv", low_memory=False)
print("df2023 shape:", df2023.shape)
display(df2023.head())
```

```python
# Get info on columns and data types
df2020.info()
df2023.info()

# Summary statistics for numeric columns
df2020.describe()
df2023.describe()
```

```python
#Let's make sure the data is clean and ready for analysis.
#1. Make sure we don't have rows where price is missing

# Show all rows where 'price' is NaN
missing_price_2020 = df2020[df2020["price"].isna()]
missing_price_2023 = df2023[df2023["price"].isna()]

missing_price_2020.head()
missing_price_2023.head()
```

```python
#Now that we're sure price is all present in all rows on both our csv, we will decide to either use neighborhood or neighborhood_group
#Check missing values on both columns
df2020[["neighbourhood_group", "neighbourhood"]].isna().sum()
```

```python
df2023[["neighbourhood_group", "neighbourhood"]].isna().sum()
```

```python
#This tells us that neighborhood_group column has a lot of missing/NaN values. Now checking for unique values for both columns.
print("Unique values in 'neighbourhood_group':", df2020["neighbourhood_group"].unique())
print("\nUnique values in 'neighbourhood':", df2020["neighbourhood"].unique())

print("Unique values in 'neighbourhood_group':", df2023["neighbourhood_group"].unique())
print("\nUnique values in 'neighbourhood':", df2023["neighbourhood"].unique())
```

```python
#We are choosing to drop the neighborhood_group as they are missing values plus we are fine with the more granular data.
df2020 = df2020.drop(columns=["neighbourhood_group"])
df2023 = df2023.drop(columns=["neighbourhood_group"])
```

```python
#We are seeing that some neighborhood are named as their Zip Code, we would like to change that into neighborhood names.
#We will use ZIP to Neighborhood data to impute the zip code values to the neighborhood names
# Load usneighborhoods CSV
zip_df = pd.read_csv("usneighborhoods.csv", delimiter=",")

# Check the first few rows
zip_df.head()
```

```python
# We see that zips may have more than one zip code so we have to normalize/explode the data into their own rows
# Ensure 'zips' is treated as a string (in case of NaN values)
zip_df["zips"] = zip_df["zips"].astype(str)

# Split 'zips' into a list (splitting by space), then explode into multiple rows
zip_expanded = zip_df.assign(zip_code=zip_df["zips"].str.split(" ")).explode("zip_code")

# Convert ZIP codes to string (to avoid integer conversion issues)
zip_expanded["zip_code"] = zip_expanded["zip_code"].astype(str)

# Keep only relevant columns for merging
zip_expanded = zip_expanded[["zip_code", "neighborhood"]]

# Display the expanded DataFrame
display(zip_expanded.head(20))  # Shows the first 20 rows
```

```python
# We might have to supplement the list with our own data as Texas data doesn't exist in the usneighborhoods.csv

# Create a new DataFrame with multiple ZIP codes and their neighborhoods
new_entries = pd.DataFrame({
    "zip_code": ["78746", "78732", "78712", "78733", "78734"],
    "neighborhood": ["West Austin", "Steiner Ranch", "University of Texas at Austin", "Westlake Austin", "Lakeway"]
})

# Append to zip_expanded
zip_expanded = pd.concat([zip_expanded, new_entries], ignore_index=True)
```

```python
#Drop duplicates
zip_expanded = zip_expanded.drop_duplicates(subset=["zip_code", "neighborhood"], keep="first").reset_index(drop=True)
```

```python
#Group by Neighborhood - 2020 Data
#Using mean and show number of listings in that neighborhood
grouped_2020 = (
    df2020
    .groupby("neighbourhood", as_index=False)
    .agg(
        mean_price=("price", "mean"),
        count_listings=("price", "size")    # or .count() 
    )
)
# Now sort by mean_price descending and show top 10 neighborhood by mean_price
grouped_2020.sort_values("mean_price", ascending=False, inplace=True)
grouped_2020.head(20)
```

```python
#Do the same but median
grouped_2020_median = (
    df2020
    .groupby("neighbourhood", as_index=False)
    .agg(
        median_price=("price", "median"),
        count_listings=("price", "size")  # or .count() if you only want non-null values
    )
)

# Sort by median price in descending order
grouped_2020_median.sort_values("median_price", ascending=False, inplace=True)

# Show top 10 neighborhoods
grouped_2020_median.head(20)
```

```python
#Group by Neighborhood - 2023 Data
#Using mean and show number of listings in that neighborhood
grouped_2023 = (
    df2023
    .groupby("neighbourhood", as_index=False)
    .agg(
        mean_price=("price", "mean"),
        count_listings=("price", "size")    # or .count() 
    )
)
# Now sort by mean_price descending and show top 10 neighborhood by mean_price
grouped_2023.sort_values("mean_price", ascending=False, inplace=True)
grouped_2023.head(20)
```

```python
#Do the same but median for 2023
grouped_2023_median = (
    df2023
    .groupby("neighbourhood", as_index=False)
    .agg(
        median_price=("price", "median"),
        count_listings=("price", "size")  # or .count() if you only want non-null values
    )
)

# Sort by median price in descending order
grouped_2023_median.sort_values("median_price", ascending=False, inplace=True)

# Show top 10 neighborhoods
grouped_2023_median.head(20)
```

```python
#By doing top 20 data for both years and both median and average price, we can have a better sense of the data. Also we are choosing the median data as apparently the average rank may be skewed by a really high listing of one
#such as in West Compton neighborhood
#Let's do a comparison check, by now selecting the top 50 of each year and merge them together
top50_2020 = grouped_2020_median.head(50)
top50_2023 = grouped_2023_median.head(50)
#merge
comparison_df = (
    top50_2020
    .merge(top50_2023, on="neighbourhood", suffixes=("_2020", "_2023"), how="outer")
)

# Fill NaN if a neighborhood appears in only one dataset
comparison_df.fillna(0, inplace=True)

# Display the first 20 rows
comparison_df.head(20)
```

```python
# Now we are seeing the zip codes on the neighbourhood so we need to impute the zip codes in this column in comparison_df with actual names
# Ensure 'neighbourhood' in comparison_df is a string (to match zip_code format)
comparison_df["neighbourhood"] = comparison_df["neighbourhood"].astype(str)

# Merge only where 'neighbourhood' contains ZIP codes
comparison_df = comparison_df.merge(zip_expanded, left_on="neighbourhood", right_on="zip_code", how="left")

# Replace ZIP codes with actual neighborhood names where available
comparison_df["neighbourhood"] = comparison_df["neighborhood"].fillna(comparison_df["neighbourhood"])

# Drop unnecessary columns
comparison_df.drop(columns=["zip_code", "neighborhood"], inplace=True)
```

```python
#Let's try sorting by highest median price in 2023
comparison_df.sort_values("median_price_2023", ascending=False, inplace=True)
comparison_df.head(20)
```

```python
#Now have a check on biggest price change (signifying a glow up of the neighborhood in terms of airbnb)
# Fill NaN values (in case a neighborhood is in one year but not the other)
comparison_df.fillna(0, inplace=True)

# Recalculate percent change (AFTER merge to ensure it exists)
comparison_df["price_change"] = comparison_df["median_price_2023"] - comparison_df["median_price_2020"]
comparison_df["percent_change"] = (comparison_df["price_change"] / comparison_df["median_price_2020"]) * 100

# Now create a filtered version that excludes `inf` and `-100`
comparison_df_filtered = comparison_df.copy()

# Replace inf values and -100% changes with NaN
comparison_df_filtered["percent_change"] = comparison_df_filtered["percent_change"].replace([np.inf, -np.inf, -100], np.nan)

# Drop rows where percent_change is NaN
comparison_df_filtered = comparison_df_filtered.dropna(subset=["percent_change"])

# Sort by biggest percentage increase
comparison_df_filtered = comparison_df_filtered.sort_values("percent_change", ascending=False)

# Display top 20 neighborhoods with biggest valid increase
comparison_df_filtered.head(20)
```

```python
# Now that we have answered question no 1, time to do our question no 2
# Analyze total number of listings in 2020 and 2023 based on open data that we have
total_listings_2020 = df2020["id"].nunique()
total_listings_2023 = df2023["id"].nunique()

print(f"Total Listings in 2020: {total_listings_2020}")
print(f"Total Listings in 2023: {total_listings_2023}")

# Percentage growth
growth_percentage = ((total_listings_2023 - total_listings_2020) / total_listings_2020) * 100
print(f"Percentage Growth in Listings: {growth_percentage:.2f}%")
```

```python
# Number of New Hosts in 2023
hosts_2020 = set(df2020["host_id"].unique())
hosts_2023 = set(df2023["host_id"].unique())

new_hosts_2023 = hosts_2023 - hosts_2020  # Hosts who joined in 2023
num_new_hosts_2023 = len(new_hosts_2023)

# Percentage of new hosts in 2023 relative to total hosts in 2023
total_hosts_2023 = len(hosts_2023)
new_hosts_percentage = (num_new_hosts_2023 / total_hosts_2023) * 100

print(f"Number of New Hosts in 2023: {num_new_hosts_2023}")
print(f"Percentage of 2023 Hosts That Are New: {new_hosts_percentage:.2f}%")
```

```python
# Number of Old Hosts that Left
# Find hosts who were in 2020 but not in 2023
old_hosts = hosts_2020 - hosts_2023
num_old_hosts = len(old_hosts)

print(f"Number of Hosts in 2020 who are NOT in 2023: {num_old_hosts}")

total_hosts_2020 = len(hosts_2020)

old_hosts_percentage = (num_old_hosts / total_hosts_2020) * 100

print(f"Percentage of 2020 Hosts Who Left Airbnb by 2023: {old_hosts_percentage:.2f}%")
```

```python
# Compute Average Listings Per Host
avg_listings_per_host_2020 = df2020["calculated_host_listings_count"].mean()
avg_listings_per_host_2023 = df2023["calculated_host_listings_count"].mean()

print(f"Average Listings per Host in 2020: {avg_listings_per_host_2020:.2f}")
print(f"Average Listings per Host in 2023: {avg_listings_per_host_2023:.2f}")

# Percentage growth in multi-listing hosts
multi_host_growth = ((avg_listings_per_host_2023 - avg_listings_per_host_2020) / avg_listings_per_host_2020) * 100
print(f"Percentage Change in Multi-Listing Hosts: {multi_host_growth:.2f}%")
```

```python
# Now let's create some data visualization to be put in the blog post
import matplotlib.pyplot as plt

# Total Listings Growth (Bar Chart)
years = ["2020", "2023"]
listings = [total_listings_2020, total_listings_2023]

plt.figure(figsize=(6, 4))
plt.bar(years, listings, color=["blue", "green"])
plt.xlabel("Year")
plt.ylabel("Total Listings")
plt.title("Total Airbnb Listings (2020 vs 2023)")
plt.show()

# Average Listings per Host (Bar Chart)
avg_listings = [avg_listings_per_host_2020, avg_listings_per_host_2023]

plt.figure(figsize=(6, 4))
plt.bar(years, avg_listings, color=["blue", "green"])
plt.xlabel("Year")
plt.ylabel("Average Listings per Host")
plt.title("Average Number of Listings Per Host (2020 vs 2023)")
plt.show()

# New Hosts as a Percentage of 2023 Hosts (Pie Chart)
labels = ["New Hosts", "Existing Hosts"]
sizes = [new_hosts_percentage, 100 - new_hosts_percentage]
colors = ["gold", "lightblue"]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90)
plt.title("Percentage of New Hosts in 2023")
plt.show()

# Old Hosts Who Left (Pie Chart)
labels = ["Hosts Who Left", "Hosts Who Stayed"]
sizes = [old_hosts_percentage, 100 - old_hosts_percentage]
colors = ["red", "lightblue"]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90)
plt.title("Percentage of 2020 Hosts Who Left by 2023")
plt.show()
```
